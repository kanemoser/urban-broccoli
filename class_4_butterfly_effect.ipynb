# Butterfly Effect, aka sensitivity to initial conditions
Show the big impact that a small variation at the beginning has further down the line.

### Exercise 1: 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 100
x = np.zeros(nsteps)

starting_points = [0.6, 0.6001]
fig = plt.figure(figsize=(14,7))

for x[0] in starting_points:
    r = 3.6
    for i in range(1,nsteps):
        x[i] = r*x[i-1]*(1-x[i-1])
    plt.plot(x);
plt.ylim((0,1));
```

```python
nsteps = 100
ntraces = 20
x = np.zeros(shape=(nsteps,ntraces))
wiggle = 0.0001                           # standard deviation of the original noise in X0
r = 3.9
for itrace in range(ntraces):
    x[0,itrace] = 0.6+np.random.randn(1)*wiggle
    for istep in range(1,nsteps):
        x[istep, itrace] = r*x[istep-1,itrace]*(1-x[istep-1,itrace])
fig = plt.figure(figsize=(14,7))
plt.plot(x);
plt.ylim((0.1));
plt.xlabel('Time');
plt.ylabel('All trajectories');
```

Each column in this setup is a 'trajectory,' or one of the lines plotted in Exercise 1.

We want to get standard deviation for all trajectories at each point in time.

```python
nsteps = 100
ntraces = 20
x = np.zeros(shape=(nsteps,ntraces))
wiggle = 0.0001                           # standard deviation of the original noise in X0
r = 3.9
for itrace in range(ntraces):
    x[0,itrace] = 0.6+np.random.randn(1)*wiggle
    for istep in range(1,nsteps):
        x[istep, itrace] = r*x[istep-1,itrace]*(1-x[istep-1,itrace])
fig = plt.figure(figsize=(14,7))
plt.plot(np.std(x,axis=1));
plt.xlabel('Time');
plt.ylabel('Variability (sd) of trajectories');
```
