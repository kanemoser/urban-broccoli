# Logistic map bifurcation diagram
This model illustrates chaotic behavior.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 50
r = 3
fig = plt.figure(figsize=((8,8)))
x = .67
i=0
for r in np.arange(2.6,4,0.0002):
    i=i+1
    for j in range(nsteps+i%50): # must use 
        x = r*x*(1-x)
    plt.plot(r,x,'k.',ms=1.5)
```

* This thing is conceptually random but factually deterministic
* We performed a parameter sweep, which is building all possible parameter values within a range and seeing what happens within the model (this is better than manually changing value and seeing what happens)
