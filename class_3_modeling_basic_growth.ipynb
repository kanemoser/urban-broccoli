# Growth
### Exponential model: Bacteria in the Broth
We want to determine changes in the number of bacteria present, since we can assume they replicate.

$t+\Delta t=x*r$   This is an honest equation.

$x_{i+1}=x*r$   This is the equation we will use.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 100             # Number of steps in our simulation
x = np.zeros(nsteps)     # Number of animals at every step
x[0] = 1                 # The original number of animals
r = 1.1                   # Rate of population change (growth - die out, sort of)

for i in range(1,nsteps):
    x[i] = x[i-1]*r

plt.plot(x);
```

$newN=n+\Delta n=n-en+kn=n+(k-e)n$

What is the simplest formula we could use?

$x=\frac{n}{nMax}$

$x=1$: on target

$x>1$: too many

$x<1$: too few

$x_{i+1}=x_{i}+bx_{i}(1-x_{i})$

Simplified:

$x(1+b-bx)$

Simplified further:

$x_{i}=r*x(1-x_{i-1})$


We are not actually thinking about the number of animals, we are thinking about how far we are away from the carrying capacity for the population.

```python
nsteps = 100             # Number of steps in our simulation
x = np.zeros(nsteps)     # Number of animals at every step
x[0] = 0.6               # The original population size
r = 3.6                    # Rate of population change (growth - die out, sort of like speed of reaction)

for i in range(1,nsteps):
    x[i] = r*x[i-1]*(1-x[i-1])

plt.plot(x);
plt.ylim((0,1)) # force axes to be between 0 and 1
```

* If r is high, the population cannot actually reach stable point; it always overshoots one way or the other, ocillating around the point of stability.
