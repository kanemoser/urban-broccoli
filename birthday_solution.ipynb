# Solution to the Birthday Problem

```python
import numpy as np
import matplotlib.pyplot as plt
```

The logic:

* People are showing up one by one (we pretend that they are showing up one by one, as it makes thinking about this easier).
* Once n-th person comes into the room, there are n-1 people already there. Let's calculate P(that there is no overlap in birthdays). Each person claims a day of the year.
* With n-1 person, this means n-1 days taken. (n-1)/365 is taken.
* The probability of a new person not falling into the "taken" portion of the year is therefore 1-(n-1)/365
* So we have enough to calculate it. Each new n-th person maintains the status quo of "no overlap in birthdays with 1-(n-1)/3665 probability. And originally, for n-1, P(no overlap)=1

```python
maxn=80          # Highest number of people we'll consider
p=np.zeros(maxn)  # Future answer, p(n). We'll create an array and fill it with zeros
p[0]=1            # If no people, p=1
p[1]=1            # If 1 person, also no overlap
for i in range(2,maxn):
    p[i]=p[i-1]*(1-(i-1)/365)
    
plt.plot(1-p);
plt.xlabel('N people in the group');
plt.ylabel('P of shared birthday');
```
