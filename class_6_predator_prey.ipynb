# Predator-Prey Interaction Model
Get the basic model running along with a phase portrait plot:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 100000
a = .6 # bunny reproduction rate
b = 1.2 # bunny death
c = .5 # fox reproduction rate
d = .1 # fox death
dt = 0.001 # step

x = np.zeros(nsteps)
y = np.zeros(nsteps)
x[0] = 1 # initial bunnies
y[0] = 1 # initial foxes

for t in range(1,nsteps):
    x[t] = x[t-1] + dt*(a*x[t-1] - b*x[t-1]*y[t-1])
    y[t] = y[t-1] + dt*(c*x[t-1]*y[t-1] - d*y[t-1])

fig = plt.figure() # Number of individuals in each species over time
plt.plot(x, label='Bunnies');
plt.plot(y, label='Foxes');
plt.legend(loc='upper right');
plt.xlabel('Time')
plt.ylabel('Species')

fig = plt.figure(figsize=(5,5)) # Phase portrait plot
plt.plot(x,y);
plt.xlabel('Bunnies');
plt.ylabel('Foxes');
```
